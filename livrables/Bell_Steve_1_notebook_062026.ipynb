```python
# Fonctions d’affichage simples et génération des exports HTML


def _formatter_colonne(nom_colonne: str, serie: pd.Series):
    """Détermine un format d'affichage adapté au nom et au contenu d'une colonne."""
    col = str(nom_colonne).lower()

    if not pd.api.types.is_numeric_dtype(serie):
        return None

    format_nombre_2 = lambda x: "" if pd.isna(x) else f"{x:,.2f}".replace(",", " ").replace(".", ",")
    format_pourcentage_0_100 = lambda x: "" if pd.isna(x) else f"{x:,.2f} %".replace(",", " ").replace(".", ",")
    format_pourcentage_0_1 = lambda x: "" if pd.isna(x) else f"{x:,.2%}".replace(".", ",")
    format_euros = lambda x: "" if pd.isna(x) else f"{x:,.2f} €".replace(",", " ").replace(".", ",")
    format_entier = lambda x: "" if pd.isna(x) else f"{x:,.0f}".replace(",", " ")

    # Règles spécifiques à appliquer avant les règles monétaires : z_score_price ne doit pas être lu comme un prix.
    if col in ["z_score", "z_score_price"] or col.startswith("z_score"):
        return format_nombre_2

    if col == "mois_stock" or "mois" in col:
        return lambda x: "" if pd.isna(x) else f"{x:,.2f} mois".replace(",", " ").replace(".", ",")

    if any(mot in col for mot in ["corrélation", "correlation", "corr"]):
        return format_nombre_2

    if "taux_marge" in col:
        return format_pourcentage_0_100

    if col == "part_ca" or col == "part_quantite" or col.startswith("part_"):
        return format_pourcentage_0_1

    # Identifiants : éviter les décimales et les symboles monétaires.
    if any(mot in col for mot in ["id", "sku", "key"]):
        return format_entier

    # Colonnes déjà exprimées en pourcentage 0-100.
    if "%" in col:
        return format_pourcentage_0_100

    # Ratios / taux en général stockés entre 0 et 1, hors taux_marge traité ci-dessus.
    if any(mot in col for mot in ["taux", "ratio"]):
        return format_pourcentage_0_1

    # Montants monétaires explicites. Attention : z_score_price est traité plus haut.
    colonnes_euros = ["price", "prix", "purchase_price", "ca_par_article", "marge_unitaire", "marge_totale"]
    if col in colonnes_euros or col.startswith("valorisation_") or any(mot in col for mot in ["chiffre", "achat"]):
        return format_euros

    # Volumes, comptes, stocks, ventes.
    if any(mot in col for mot in ["stock_quantity", "total_sales", "nombre", "lignes", "colonnes", "valeurs", "quantite", "quantité", "ventes"]):
        return format_entier

    return format_nombre_2


def _slugifier(texte: str) -> str:
    """Convertit un titre en nom de fichier simple."""
    texte = str(texte).strip().lower()
    texte = re.sub(r"[^a-z0-9A-ZÀ-ÿ]+", "_", texte)
    texte = texte.strip("_")
    return texte[:80] or "tableau"


_EXPORT_TABLEAU_COMPTEUR = 0
_EXPORTS_HTML_GENERES: list[dict] = []
_EXPORT_TITRES_UTILISES: dict[str, int] = {}
_EXPORT_HASH_CONTENU: dict[str, Path] = {}
_EXPORT_DOUBLONS_IGNORES: list[dict] = []
_EXPORTS_HTML_PREPARES = False


def preparer_exports_html(force_clean: bool = True) -> Path:
    """Prépare le dossier des exports HTML et réinitialise les registres de l'exécution courante.

    Le nettoyage est strictement limité aux fichiers ``*.html`` présents directement dans
    ``exports/tableaux`` afin de ne jamais toucher au notebook ni aux autres livrables.
    """
    global _EXPORT_TABLEAU_COMPTEUR, _EXPORTS_HTML_GENERES, _EXPORT_TITRES_UTILISES
    global _EXPORT_HASH_CONTENU, _EXPORT_DOUBLONS_IGNORES, _EXPORTS_HTML_PREPARES

    dossier_export = Path("exports") / "tableaux"
    dossier_export.mkdir(parents=True, exist_ok=True)
    dossier_resolu = dossier_export.resolve()
    dossier_attendu = (Path.cwd() / "exports" / "tableaux").resolve()
    if dossier_resolu != dossier_attendu or dossier_resolu.name != "tableaux" or dossier_resolu.parent.name != "exports":
        raise RuntimeError(f"Nettoyage HTML refusé hors du dossier d'exports attendu : {dossier_resolu}")

    if force_clean:
        for ancien_export in dossier_export.glob("*.html"):
            ancien_resolu = ancien_export.resolve()
            if ancien_resolu.parent != dossier_resolu or ancien_export.suffix.lower() != ".html":
                raise RuntimeError(f"Suppression HTML refusée pour un chemin non sécurisé : {ancien_resolu}")
            ancien_export.unlink()

    _EXPORT_TABLEAU_COMPTEUR = 0
    _EXPORTS_HTML_GENERES = []
    _EXPORT_TITRES_UTILISES = {}
    _EXPORT_HASH_CONTENU = {}
    _EXPORT_DOUBLONS_IGNORES = []
    _EXPORTS_HTML_PREPARES = True
    return dossier_export


def nettoyer_exports_tableaux() -> None:
    """Compatibilité : délègue au préparateur centralisé des exports HTML."""
    preparer_exports_html(force_clean=True)


def _hash_dataframe_export(df: pd.DataFrame) -> str:
    """Calcule une empreinte stable du contenu réellement exporté."""
    contenu = pd.util.hash_pandas_object(df.astype(str).fillna(""), index=True).values.tobytes()
    colonnes = "||".join(map(str, df.columns)).encode("utf-8")
    return hashlib.sha256(colonnes + b"\0" + contenu).hexdigest()


def _libelle_index_export(titre: str, nom_fichier: str, df: pd.DataFrame) -> str:
    """Améliore les libellés génériques affichés dans l'index sans filtrer par titre."""
    nom = _slugifier(nom_fichier).lower()
    cols = {str(col).lower().strip() for col in df.columns}
    texte = f"{nom} {' '.join(sorted(cols))} {titre.lower()}"
    mapping = [
        (["dashboard"], "Accueil / Dashboard"),
        (["controle_des_fichiers_charges"], "Contrôle des fichiers chargés"),
        (["base_consolidee", "df_analyse"], "Aperçu de la base consolidée"),
        (["journal", "anomal"], "Journal des anomalies"),
        (["ca_total"], "CA total — synthèse"),
        (["ca_par_article", "top"], "CA par produit — top références"),
        (["part_ca_cumulee"], "CA cumulé — analyse 20/80"),
        (["total_sales", "sku"], "Ventes par SKU — contrôle"),
        (["part_quantite_cumulee"], "Quantités cumulées — analyse 20/80"),
        (["iqr"], "Prix atypiques — méthode IQR"),
        (["z_score"], "Prix atypiques — z-score"),
        (["stock_quantity", "mois_stock"], "Stocks — couverture en mois"),
        (["valorisation_stock"], "Stocks — valorisation"),
        (["stock"], "Stocks — contrôle"),
        (["taux_marge", "marge_totale"], "Marges — synthèse"),
        (["marge"], "Marges — contrôle"),
        (["corrélation"], "Corrélations — variables clés"),
        (["correlation"], "Corrélations — variables clés"),
        (["recommandation"], "Recommandations ERP"),
        (["valeurs_manquantes", "erp"], "Valeurs manquantes ERP"),
        (["valeurs_manquantes", "web"], "Valeurs manquantes Web"),
        (["valeurs_manquantes", "liaison"], "Valeurs manquantes Liaison"),
        (["colonne", "type", "erp"], "Types de colonnes ERP"),
        (["colonne", "type", "web"], "Types de colonnes Web"),
        (["colonne", "type", "liaison"], "Types de colonnes Liaison"),
    ]
    for mots, libelle in mapping:
        if all(mot in texte for mot in mots):
            return libelle
    generiques = ["analyse du chiffre d'affaires", "tableau —", "tableau exporte", "tableau exporté"]
    if any(titre.lower().startswith(g) for g in generiques):
        return titre.replace("Tableau — ", "").replace("Tableau exporté", "Tableau détaillé")
    return titre


def _enregistrer_export_html(chemin: Path, titre: str, label: str, categorie: str, description: str, hash_contenu: str | None = None) -> None:
    """Ajoute un export créé pendant l'exécution courante au registre utilisé par l'index."""
    global _EXPORTS_HTML_GENERES, _EXPORT_TITRES_UTILISES
    compteur_titre = _EXPORT_TITRES_UTILISES.get(label, 0) + 1
    _EXPORT_TITRES_UTILISES[label] = compteur_titre
    label_index = f"{label} — vue {compteur_titre}" if compteur_titre > 1 else label
    _EXPORTS_HTML_GENERES.append({
        "path": chemin,
        "titre": titre,
        "label": label_index,
        "categorie": categorie,
        "description": description,
        "hash": hash_contenu,
    })


def _categorie_export(label: str, nom_fichier: str) -> str:
    texte = f"{label} {nom_fichier}".lower()
    mots_cles = ["dashboard", "base consolidée", "fichiers chargés", "qualité", "journal", "anomal", "ca ", "chiffre", "top", "20/80", "prix atyp", "iqr", "stock", "marge", "corr", "recommand"]
    return "Exports clés" if any(mot in texte for mot in mots_cles) else "Exports techniques détaillés"


def _description_export(label: str, nom_fichier: str) -> str:
    texte = f"{label} {nom_fichier}".lower()
    if "dashboard" in texte:
        return "Vue exécutive avec KPI, graphiques et tableaux prioritaires."
    if "ca" in texte or "chiffre" in texte:
        return "Analyse du chiffre d'affaires et de sa contribution par produit."
    if "stock" in texte:
        return "Suivi des stocks, disponibilités ou valorisations."
    if "marge" in texte:
        return "Analyse de marge et contrôles de rentabilité."
    if "anomal" in texte or "qualité" in texte or "controle" in texte or "contrôle" in texte:
        return "Contrôle de fiabilité des données et points à corriger."
    if "valeurs manquantes" in texte or "types de colonnes" in texte:
        return "Contrôle technique utile pour tracer la qualité des fichiers sources."
    return "Export HTML consultable dans l'aperçu central."


preparer_exports_html(force_clean=True)


def _nom_export_tableau(titre: str | None = None, nom_fichier: str | None = None) -> str:
    """Retourne un nom HTML stable et lisible sans écraser les tableaux non titrés."""
    global _EXPORT_TABLEAU_COMPTEUR

    if nom_fichier:
        return nom_fichier
    if titre:
        return titre

    _EXPORT_TABLEAU_COMPTEUR += 1
    return f"tableau_exporte_{_EXPORT_TABLEAU_COMPTEUR:03d}"


def _titre_depuis_colonnes(df: pd.DataFrame) -> str:
    """Déduit un titre lisible à partir des colonnes quand aucun titre explicite n'est fourni."""
    colonnes = [str(col) for col in df.columns]
    normalisees = {col.lower().strip() for col in colonnes}
    colonnes_concat = " ".join(sorted(normalisees))

    if {"product_id", "sku", "post_title", "price", "total_sales"}.issubset(normalisees):
        return "Top produits par prix et ventes"
    if {"product_id", "sku", "post_title", "stock_quantity"}.issubset(normalisees):
        return "Suivi des stocks par produit"
    if {"product_id", "sku", "price"}.issubset(normalisees):
        return "Contrôle des prix par produit"
    if {"sku", "total_sales"}.issubset(normalisees):
        return "Analyse des ventes par SKU"
    if {"sku", "product_id"}.issubset(normalisees):
        return "Correspondance produits et SKU"
    if any(mot in colonnes_concat for mot in ["anomal", "alerte", "erreur", "incoher", "incohér"]):
        return "Contrôle des anomalies détectées"
    if any(mot in colonnes_concat for mot in ["marge", "profit"]):
        return "Analyse des marges"
    if any(mot in colonnes_concat for mot in ["ca", "chiffre", "revenue"]):
        return "Analyse du chiffre d'affaires"

    principales = [col.replace("_", " ") for col in colonnes[:4]]
    if principales:
        return "Tableau — " + ", ".join(principales).title()
    return "Tableau exporté"


def _titre_export_tableau(titre: str | None = None, df: pd.DataFrame | None = None) -> str:
    """Produit un titre compréhensible pour les tableaux exportés sans titre explicite."""
    if titre:
        return titre
    if df is not None:
        return _titre_depuis_colonnes(df)
    return "Tableau exporté"


def _preparer_html_tableau(df: pd.DataFrame, masquer_index: bool | None = None) -> str:
    """Prépare le HTML d'un DataFrame avec formatage des colonnes."""
    if masquer_index is None:
        masquer_index = isinstance(df.index, pd.RangeIndex)

    formatters = {
        col: fmt for col in df.columns
        if (fmt := _formatter_colonne(str(col), df[col])) is not None
    }

    return df.to_html(
        classes="bn-table",
        index=not masquer_index,
        border=0,
        escape=True,
        formatters=formatters,
        na_rep=""
    )


def exporter_tableau_html(obj, titre: str = "Tableau", nom_fichier: str | None = None, max_lignes: int | None = None) -> Path | None:
    """Exporte un tableau dans un fichier HTML local lisible dans le navigateur."""
    if isinstance(obj, pd.Series):
        obj = obj.reset_index()

    if not isinstance(obj, pd.DataFrame):
        return None

    df = obj.copy()
    if max_lignes is not None:
        df = df.head(max_lignes)

    global _EXPORTS_HTML_PREPARES, _EXPORT_HASH_CONTENU, _EXPORT_DOUBLONS_IGNORES
    if not _EXPORTS_HTML_PREPARES:
        preparer_exports_html(force_clean=True)

    dossier_export = Path("exports") / "tableaux"
    dossier_export.mkdir(parents=True, exist_ok=True)

    titre_affiche = _titre_export_tableau(titre, df)
    nom = _slugifier(nom_fichier or titre_affiche)
    hash_contenu = _hash_dataframe_export(df)
    if hash_contenu in _EXPORT_HASH_CONTENU:
        chemin_existant = _EXPORT_HASH_CONTENU[hash_contenu]
        _EXPORT_DOUBLONS_IGNORES.append({"titre": titre_affiche, "fichier_existant": chemin_existant})
        print(f"Export HTML ignoré (doublon exact déjà généré) : {titre_affiche} -> {chemin_existant}")
        return chemin_existant

    chemin = dossier_export / f"{nom}.html"
    suffixe = 2
    while chemin.exists():
        chemin = dossier_export / f"{nom}_{suffixe:02d}.html"
        suffixe += 1
    table_html = _preparer_html_tableau(df)
    date_generation = datetime.now().strftime('%d/%m/%Y %H:%M')
    nb_lignes, nb_colonnes = df.shape
    colonnes_resume = ", ".join(map(str, df.columns[:6])) + ("…" if nb_colonnes > 6 else "")

    page_html = f"""
<!doctype html>
<html lang="fr">
<head>
<meta charset="utf-8">
<meta name="viewport" content="width=device-width, initial-scale=1">
<title>{html.escape(titre_affiche)}</title>
<style>
    :root {{
        --primary:#33A5B6; --primary-dark:#1E7D8B; --ink:#17202A; --muted:#607D8B;
        --line:#D9EEF2; --soft:#F4FAFB; --white:#fff; --shadow:0 18px 42px rgba(17,24,39,.10);
    }}
    * {{ box-sizing: border-box; }}
    html {{ scroll-behavior: smooth; }}
    body {{
        margin: 0;
        font-family: Inter, Segoe UI, Arial, sans-serif;
        color: var(--ink);
        background: radial-gradient(circle at top left, #EAF9FC 0, #F7FBFC 34%, #EEF4F8 100%);
    }}
    .hero {{
        padding: 28px clamp(18px, 4vw, 48px);
        color: white;
        background: linear-gradient(135deg, #33A5B6 0%, #1E7D8B 55%, #111827 100%);
    }}
    .hero h1 {{ margin: 0 0 10px; font-size: clamp(1.7rem, 3.5vw, 2.8rem); line-height: 1.05; }}
    .hero p {{ margin: 0; max-width: 960px; color: #EAF9FC; line-height: 1.55; }}
    main {{ padding: 20px clamp(14px, 3vw, 40px) 36px; }}
    .summary {{ display: grid; grid-template-columns: repeat(auto-fit, minmax(190px, 1fr)); gap: 12px; margin-bottom: 16px; }}
    .card {{ background: var(--white); border: 1px solid var(--line); border-radius: 18px; padding: 14px 16px; box-shadow: var(--shadow); }}
    .card span {{ display:block; color:var(--muted); font-size:.78rem; text-transform:uppercase; letter-spacing:.07em; font-weight:800; }}
    .card strong {{ display:block; margin-top:6px; font-size:1.25rem; color:#111827; }}
    .toolbar {{
        position: sticky; top: 0; z-index: 5; display:flex; gap:12px; flex-wrap:wrap; align-items:center;
        background: rgba(247,251,252,.92); backdrop-filter: blur(10px); border:1px solid var(--line);
        border-radius: 18px; padding: 12px; margin-bottom: 16px; box-shadow: 0 10px 24px rgba(17,24,39,.07);
    }}
    .toolbar input {{ flex: 1 1 320px; min-width: 220px; padding: 12px 14px; border: 1px solid #B7DDE4; border-radius: 12px; font-size: 15px; outline: none; background:white; }}
    .toolbar input:focus {{ border-color: var(--primary); box-shadow: 0 0 0 4px rgba(51,165,182,.16); }}
    .counter {{ color: var(--muted); font-weight: 800; }}
    .table-wrap {{ border: 1px solid var(--line); background: white; border-radius: 18px; overflow: auto; max-height: 76vh; box-shadow: var(--shadow); }}
    table.bn-table {{ border-collapse: separate; border-spacing: 0; width: max-content; min-width: 100%; font-size: 13px; }}
    table.bn-table thead th {{ position: sticky; top: 0; z-index: 3; background: #1E7D8B; color: white; text-align: left; padding: 11px 13px; border-bottom: 1px solid #166A75; white-space: nowrap; }}
    table.bn-table tbody td, table.bn-table tbody th {{ padding: 9px 13px; border-bottom: 1px solid #E5EEF0; vertical-align: top; white-space: nowrap; }}
    table.bn-table tbody tr:nth-child(even) {{ background: #F8FBFC; }}
    table.bn-table tbody tr:hover {{ background: #EAF9FC; }}
</style>
</head>
<body>
<header class="hero">
    <h1>{html.escape(titre_affiche)}</h1>
    <p>Tableau HTML généré automatiquement par le notebook BottleNeck, avec recherche intégrée et en-têtes figés pour faciliter la lecture.</p>
</header>
<main>
    <section class="summary" aria-label="Résumé du tableau">
        <div class="card"><span>Lignes affichées</span><strong>{nb_lignes}</strong></div>
        <div class="card"><span>Colonnes</span><strong>{nb_colonnes}</strong></div>
        <div class="card"><span>Génération</span><strong>{date_generation}</strong></div>
        <div class="card"><span>Colonnes principales</span><strong>{html.escape(colonnes_resume) or '-'}</strong></div>
    </section>
    <div class="toolbar">
        <input id="searchInput" type="search" placeholder="Rechercher dans ce tableau..." aria-label="Rechercher dans ce tableau">
        <span class="counter" id="rowCounter">{nb_lignes} ligne(s)</span>
    </div>
    <div class="table-wrap">
        {table_html}
    </div>
</main>
<script>
const input = document.getElementById('searchInput');
const counter = document.getElementById('rowCounter');
const rows = Array.from(document.querySelectorAll('table.bn-table tbody tr'));
function filterTable() {{
    const filter = input.value.trim().toLowerCase();
    let visible = 0;
    rows.forEach(row => {{
        const ok = row.textContent.toLowerCase().includes(filter);
        row.style.display = ok ? '' : 'none';
        if (ok) visible += 1;
    }});
    counter.textContent = visible + ' ligne(s) visible(s)';
}}
input.addEventListener('input', filterTable);
</script>
</body>
</html>
"""

    chemin.write_text(page_html, encoding="utf-8")
    _EXPORT_HASH_CONTENU[hash_contenu] = chemin
    label = _libelle_index_export(titre_affiche, nom, df)
    categorie = _categorie_export(label, nom)
    description = _description_export(label, nom)
    _enregistrer_export_html(chemin, titre_affiche, label, categorie, description, hash_contenu)
    return chemin


def afficher_tableau(obj, titre: str | None = None, max_lignes: int | None = None, masquer_index: bool | None = None,
                     nom_fichier: str | None = None):
    """Affiche un tableau dans le notebook et exporte une version HTML, sans renvoyer le DataFrame.

    Important : dans Jupyter/VS Code, la valeur renvoyée par la dernière instruction d'une cellule
    est affichée automatiquement. Si cette fonction renvoie le DataFrame après display(df), le
    notebook montre donc deux tableaux : l'affichage explicite puis le résultat de retour.
    """
    if isinstance(obj, pd.Series):
        obj = obj.reset_index()

    if not isinstance(obj, pd.DataFrame):
        display(obj)
        return None

    df = obj.copy()
    if max_lignes is not None:
        df = df.head(max_lignes)

    chemin_html = None
    if not df.empty or titre or nom_fichier:
        chemin_html = exporter_tableau_html(
            df,
            titre=_titre_export_tableau(titre, df),
            nom_fichier=_nom_export_tableau(titre=titre, nom_fichier=nom_fichier)
        )

    if titre:
        print()
        print(titre)
        print("-" * len(titre))
    if chemin_html is not None:
        print(f"Export HTML : {chemin_html}")

    display(df)
    return None


def afficher_kpis(kpis: list[dict]):
    """Affiche les KPI sous forme de tableau simple compatible Colab."""
    tableau_kpis = pd.DataFrame(kpis)
    colonnes = [col for col in ["label", "value", "note"] if col in tableau_kpis.columns]
    if colonnes:
        tableau_kpis = tableau_kpis[colonnes]
    display(tableau_kpis)
    return None


def afficher_message(message: str, type_message: str = "info"):
    """Affiche un message texte simple dans le notebook."""
    prefixe = {
        "info": "Information",
        "success": "Succès",
        "warning": "Attention",
        "danger": "Alerte",
    }.get(type_message, "Information")
    print(f"{prefixe} : {message}")
```

```python
# ============================================================
# Recherche automatique du dossier contenant les fichiers data
# ============================================================

fichiers_attendus = ["erp.xlsx", "web.xlsx", "liaison.xlsx"]

def contient_fichiers_attendus(dossier: Path) -> bool:
    """Vérifie si un dossier contient tous les fichiers nécessaires."""
    return dossier.exists() and all((dossier / fichier).exists() for fichier in fichiers_attendus)


def trouver_dossier_data() -> Path:
    """
    Recherche automatiquement le dossier contenant les fichiers :
    - erp.xlsx
    - web.xlsx
    - liaison.xlsx

    Compatible avec Google Colab et une exécution locale simple.
    """

    dossier_courant = Path.cwd()

    chemins_possibles = [
        dossier_courant / "Data+Bottleneck",
        dossier_courant / "data",
        dossier_courant,
        dossier_courant.parent / "Data+Bottleneck",
        Path("/content") / "Data+Bottleneck",
        Path("/mnt/data"),
    ]

    for dossier in chemins_possibles:
        if contient_fichiers_attendus(dossier):
            return dossier.resolve()

    # Recherche automatique dans les dossiers principaux
    racines_recherche = [
        dossier_courant,
        dossier_courant.parent,
        Path("/content"),
        Path("/mnt/data"),
    ]

    dossiers_deja_testes = set()

    for racine in racines_recherche:
        if not racine.exists():
            continue

        for fichier_erp in racine.rglob("erp.xlsx"):
            dossier = fichier_erp.parent.resolve()

            if dossier in dossiers_deja_testes:
                continue

            dossiers_deja_testes.add(dossier)

            if contient_fichiers_attendus(dossier):
                return dossier

    raise FileNotFoundError(
        "Impossible de trouver les fichiers erp.xlsx, web.xlsx et liaison.xlsx.\n\n"
        "Vérifie que les fichiers sont bien présents dans l'un de ces dossiers :\n"
        "- Data+Bottleneck/\n"
        "- data/\n"
        "- le même dossier que le notebook\n"
        "- /content/Data+Bottleneck/ dans Google Colab."
    )


DATA_DIR = trouver_dossier_data()

print("Dossier de données utilisé :")
print(DATA_DIR)

print("\nFichiers détectés :")
for fichier in fichiers_attendus:
    print(f"- {fichier} : OK")
```

```python
def resume_dimensions(df: pd.DataFrame, nom_fichier: str) -> pd.DataFrame:
    """Retourne les dimensions d'un DataFrame sous forme de tableau."""
    return pd.DataFrame({
        "Fichier": [nom_fichier],
        "Nombre de lignes": [df.shape[0]],
        "Nombre de colonnes": [df.shape[1]]
    })


def synthese_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    """Retourne une synthèse des colonnes, types, valeurs manquantes et valeurs uniques."""
    return pd.DataFrame({
        "Colonne": df.columns,
        "Type": df.dtypes.astype(str).values,
        "Valeurs non nulles": df.notna().sum().values,
        "Valeurs manquantes": df.isna().sum().values,
        "% valeurs manquantes": (df.isna().mean().values * 100).round(2),
        "Valeurs uniques": df.nunique(dropna=True).values
    })


def nettoyer_cle(valeur):
    """Nettoie une clé de jointure en conservant les identifiants sous forme de texte homogène.

    Les valeurs absentes restent de vraies valeurs manquantes (`np.nan`) afin de pouvoir
    les exclure explicitement des jointures. Cela évite les correspondances accidentelles
    entre deux clés manquantes, par exemple `id_web = NaN` côté liaison et `sku = NaN`
    côté Web.
    """
    if pd.isna(valeur):
        return np.nan

    texte = str(valeur).strip()
    if texte == "" or texte.lower() in {"nan", "none", "nat", "null"}:
        return np.nan

    # Cas fréquent : Excel lit un identifiant comme 12345.0
    if texte.endswith(".0"):
        texte = texte[:-2]
    return texte


def rapport_cles_manquantes(df: pd.DataFrame, colonne: str, nom_table: str) -> pd.DataFrame:
    """Retourne un contrôle simple du volume de clés de jointure manquantes."""
    nb_lignes = len(df)
    nb_manquantes = int(df[colonne].isna().sum()) if colonne in df.columns else nb_lignes
    return pd.DataFrame({
        "Table": [nom_table],
        "Clé contrôlée": [colonne],
        "Nombre de lignes": [nb_lignes],
        "Clés manquantes": [nb_manquantes],
        "% clés manquantes": [round(nb_manquantes / nb_lignes * 100, 2) if nb_lignes else 0]
    })


def tableau_value_counts(serie: pd.Series, colonne_resultat: str, colonne_nombre: str) -> pd.DataFrame:
    """Retourne un tableau de comptage stable avec Pandas 1.x et 2.x."""
    tableau = serie.value_counts(dropna=False).rename_axis(colonne_resultat).reset_index(name=colonne_nombre)
    return tableau


def preparer_fichier_erp(df_source: pd.DataFrame) -> pd.DataFrame:
    """Prépare le fichier ERP et crée la clé utilisée pour la jointure Liaison."""
    df_prepare = df_source.copy()

    colonnes_numeriques = ["product_id", "onsale_web", "price", "stock_quantity", "purchase_price"]
    for colonne in colonnes_numeriques:
        if colonne in df_prepare.columns:
            df_prepare[colonne] = pd.to_numeric(df_prepare[colonne], errors="coerce")

    if "stock_status" in df_prepare.columns:
        df_prepare["stock_status_original"] = df_prepare["stock_status"]
        df_prepare["stock_status_corrige"] = np.where(
            df_prepare["stock_quantity"].fillna(0) <= 0,
            "outofstock",
            "instock"
        )

    df_prepare["product_id_key"] = df_prepare["product_id"]
    return df_prepare


def preparer_fichier_liaison(df_source: pd.DataFrame) -> pd.DataFrame:
    """Prépare la table Liaison et crée les clés normalisées de jointure."""
    df_prepare = df_source.copy()
    df_prepare["product_id"] = pd.to_numeric(df_prepare["product_id"], errors="coerce")
    df_prepare["product_id_key"] = df_prepare["product_id"]
    df_prepare["id_web_key"] = df_prepare["id_web"].apply(nettoyer_cle)
    return df_prepare


def preparer_fichier_web(df_source: pd.DataFrame, colonnes_utiles: list[str] | None = None):
    """Prépare le fichier Web pour les contrôles et la jointure sur SKU.

    La fonction retourne trois DataFrames :
    - df_web_clean : lignes produit nettoyées ;
    - df_web_sans_sku_product : lignes produit exclues de la jointure car sans SKU ;
    - df_web_join : lignes produit utilisables pour la jointure.
    """
    if colonnes_utiles is None:
        colonnes_utiles = [
            "sku", "total_sales", "post_title", "post_status", "post_type", "guid",
            "post_date", "post_modified", "tax_status", "comment_count"
        ]

    colonnes_utiles = [colonne for colonne in colonnes_utiles if colonne in df_source.columns]

    df_products = df_source.copy()
    if "post_type" in df_products.columns:
        df_products = df_products[df_products["post_type"] == "product"].copy()

    df_clean = df_products[colonnes_utiles].copy()

    if "sku" in df_clean.columns:
        df_clean["sku_key"] = df_clean["sku"].apply(nettoyer_cle)
    else:
        df_clean["sku_key"] = np.nan

    if "total_sales" in df_clean.columns:
        df_clean["total_sales"] = pd.to_numeric(df_clean["total_sales"], errors="coerce")

    df_sans_sku = df_clean[df_clean["sku_key"].isna()].copy()
    df_join = df_clean[df_clean["sku_key"].notna()].copy()
    return df_clean, df_sans_sku, df_join


def afficher_statut(message: str):
    """Affiche un court message de statut dans le notebook."""
    print(message)
```

## 3. Jointures

<a id="etape-3"></a>
<section style="border:1px solid #bfdbfe; border-left:7px solid #1e3a8a; border-radius:14px; padding:16px 20px; margin:16px 0; background:#eff6ff; box-shadow:0 2px 8px rgba(15,23,42,0.08);">

<a id="section-3-1"></a>
<section style="border:1px solid #d9f99d; border-left:7px solid #365314; border-radius:14px; padding:16px 20px; margin:12px 0; background:#f7fee7; box-shadow:0 2px 8px rgba(15,23,42,0.08);">

    3.1 — Jonction du fichier ERP avec le fichier Liaison
  </h3>
  <div style="color:#1f2937; font-family:Arial, Helvetica, sans-serif; font-size:1rem;">
<p style="margin:8px 0; line-height:1.6;">Jointure sur la clé product_id afin d’associer à chaque produit ERP son identifiant Web.</p>
  </div>
</section>


```python
# Fusion des fichiers ERP et Liaison

# Sécurité d'exécution : si cette cellule est relancée seule après le chargement des fichiers,
# on recrée les tables préparées nécessaires à la jointure.
if "df_erp_clean" not in globals():
    df_erp_clean = preparer_fichier_erp(df_erp)

if "df_liaison_clean" not in globals():
    df_liaison_clean = preparer_fichier_liaison(df_liaison)

df_merge_erp_liaison = pd.merge(
    df_erp_clean,
    df_liaison_clean[["product_id_key", "id_web", "id_web_key"]],
    on="product_id_key",
    how="left",
    indicator="merge_erp_liaison"
)

print("Dimensions après fusion ERP + Liaison :")
afficher_tableau(resume_dimensions(df_merge_erp_liaison, "df_merge_erp_liaison"))

afficher_tableau(tableau_value_counts(
    df_merge_erp_liaison["merge_erp_liaison"],
    "Résultat de jointure",
    "Nombre de lignes"
))
```

```python
# Lignes ERP sans correspondance dans la table de liaison

anomalies_erp_sans_liaison = df_merge_erp_liaison[df_merge_erp_liaison["merge_erp_liaison"] == "left_only"].copy()
print(f"Nombre de lignes ERP sans correspondance Liaison : {len(anomalies_erp_sans_liaison)}")
afficher_tableau(anomalies_erp_sans_liaison.head())
```

```python
# Produits ERP présents dans liaison mais sans id_web renseigné

anomalies_liaison_sans_id_web = df_merge_erp_liaison[df_merge_erp_liaison["id_web_key"].isna()].copy()
print(f"Nombre de produits ERP sans id_web : {len(anomalies_liaison_sans_id_web)}")
afficher_tableau(anomalies_liaison_sans_id_web[["product_id", "id_web", "price", "stock_quantity", "onsale_web"]].head())
```

<a id="section-3-2"></a>
<section style="border:1px solid #d9f99d; border-left:7px solid #365314; border-radius:14px; padding:16px 20px; margin:12px 0; background:#f7fee7; box-shadow:0 2px 8px rgba(15,23,42,0.08);">
  <h3 style="margin:0 0 10px 0; color:#365314; font-family:Arial, Helvetica, sans-serif; font-weight:800;">
    3.2 — Jonction avec le fichier Web
  </h3>
  <div style="color:#1f2937; font-family:Arial, Helvetica, sans-serif; font-size:1rem;">
<p style="margin:8px 0; line-height:1.6;">Jointure entre id_web côté liaison et sku côté Web pour récupérer les ventes et les informations produits.</p>
  </div>
</section>


```python
# Fusion avec le fichier Web préparé
# La fusion utilise df_web_join, qui exclut les SKU manquants pour éviter les correspondances NaN = NaN.

# Sécurité d'exécution : si cette cellule est relancée seule après le chargement des fichiers,
# on recrée les tables préparées et la fusion ERP + Liaison nécessaires.
if "df_erp_clean" not in globals():
    df_erp_clean = preparer_fichier_erp(df_erp)

if "df_liaison_clean" not in globals():
    df_liaison_clean = preparer_fichier_liaison(df_liaison)

if "df_merge_erp_liaison" not in globals():
    df_merge_erp_liaison = pd.merge(
        df_erp_clean,
        df_liaison_clean[["product_id_key", "id_web", "id_web_key"]],
        on="product_id_key",
        how="left",
        indicator="merge_erp_liaison"
    )

if "df_web_join" not in globals():
    colonnes_utiles_web_preparation = globals().get("colonnes_web_utiles")
    df_web_clean, df_web_sans_sku_product, df_web_join = preparer_fichier_web(
        df_web,
        colonnes_utiles_web_preparation
    )

df_merge = pd.merge(
    df_merge_erp_liaison,
    df_web_join,
    left_on="id_web_key",
    right_on="sku_key",
    how="left",
    indicator="merge_web"
)

print("Dimensions après fusion ERP + Liaison + Web :")
afficher_tableau(resume_dimensions(df_merge, "df_merge"))

afficher_tableau(tableau_value_counts(
    df_merge["merge_web"],
    "Résultat de jointure",
    "Nombre de lignes"
))
```

```python
# Produits ERP/Liaison sans correspondance dans le fichier Web produit

anomalies_sans_correspondance_web = df_merge[
    (df_merge["id_web_key"].notna()) & (df_merge["merge_web"] == "left_only")
].copy()

print(f"Nombre de produits avec id_web mais sans correspondance Web produit : {len(anomalies_sans_correspondance_web)}")
afficher_tableau(anomalies_sans_correspondance_web[["product_id", "id_web", "id_web_key", "price", "stock_quantity", "onsale_web"]].head())
```

```python
# Création de la base d'analyse : produits présents comme lignes produit dans le Web
# Cette base est utilisée pour les analyses de CA, ventes, prix, stock et marge.

df_analyse = df_merge[df_merge["merge_web"] == "both"].copy()

# Sécuriser les colonnes numériques nécessaires aux calculs
for colonne in ["price", "purchase_price", "stock_quantity", "total_sales"]:
    if colonne in df_analyse.columns:
        df_analyse[colonne] = pd.to_numeric(df_analyse[colonne], errors="coerce")

# Drapeaux de qualité utilisés ensuite pour éviter les résultats NaN/inf dans les indicateurs métier.
df_analyse["donnee_prix_valide"] = df_analyse["price"].notna() & (df_analyse["price"] > 0)
df_analyse["donnee_vente_valide"] = df_analyse["total_sales"].notna() & (df_analyse["total_sales"] >= 0)
df_analyse["donnee_stock_valide"] = df_analyse["stock_quantity"].notna() & (df_analyse["stock_quantity"] >= 0)
df_analyse["donnee_ca_valide"] = df_analyse["donnee_prix_valide"] & df_analyse["donnee_vente_valide"]

print("Base consolidée d'analyse créée.")
afficher_tableau(resume_dimensions(df_analyse, "df_analyse"))
afficher_tableau(pd.DataFrame({
    "Contrôle": ["Prix valides", "Ventes valides", "Stocks valides", "Lignes exploitables pour le CA"],
    "Nombre de lignes": [
        int(df_analyse["donnee_prix_valide"].sum()),
        int(df_analyse["donnee_vente_valide"].sum()),
        int(df_analyse["donnee_stock_valide"].sum()),
        int(df_analyse["donnee_ca_valide"].sum())
    ]
}), titre="Contrôles qualité de la base d'analyse")
afficher_tableau(df_analyse.head(), titre="Aperçu de la base consolidée df_analyse", nom_fichier="apercu_base_consolidee")
```

```python
# Vérification des doublons après consolidation sur la clé product_id

nb_doublons_product_id_merge = df_analyse["product_id"].dropna().duplicated().sum()
print(f"Nombre de product_id en doublon dans la base d'analyse : {nb_doublons_product_id_merge}")

if nb_doublons_product_id_merge > 0:
    afficher_tableau(df_analyse[df_analyse["product_id"].notna() & df_analyse["product_id"].duplicated(keep=False)].sort_values("product_id"))
```

**Observation**

La base consolidée `df_analyse` contient les produits ERP rapprochés avec la table de liaison et retrouvés dans les lignes produit du fichier Web. Les lignes non rapprochées sont conservées dans des tableaux d’anomalies afin d’être documentées.

```python
# Calcul de la moyenne, de l'écart-type et du Z-score sur le prix

prix_moyen = df_analyse["price"].mean()
prix_ecart_type = df_analyse["price"].std()

df_analyse["z_score_price"] = (df_analyse["price"] - prix_moyen) / prix_ecart_type

resume_z_score = pd.DataFrame({
    "Indicateur": ["Prix moyen", "Écart-type du prix"],
    "Valeur": [prix_moyen, prix_ecart_type]
})

afficher_tableau(resume_z_score)
```

```python
# Articles dont le Z-score du prix est supérieur à 3 en valeur absolue

outliers_zscore_prix = df_analyse[df_analyse["z_score_price"].abs() > 3].copy()
outliers_zscore_prix = outliers_zscore_prix.sort_values("price", ascending=False)

print(f"Nombre d'outliers détectés avec le Z-score : {len(outliers_zscore_prix)}")
afficher_tableau(outliers_zscore_prix[[col for col in ["product_id", "sku", "post_title", "price", "z_score_price"] if col in outliers_zscore_prix.columns]])
```

<section style="border:1px solid #d9f99d; border-left:7px solid #365314; border-radius:14px; padding:16px 20px; margin:12px 0; background:#f7fee7; box-shadow:0 2px 8px rgba(15,23,42,0.08);">

  <div style="color:#1f2937; font-family:Arial, Helvetica, sans-serif; font-size:1rem;">
<p style="margin:8px 0; line-height:1.6;">La méthode IQR isole les valeurs situées au-delà de Q1 - 1,5×IQR ou Q3 + 1,5×IQR.</p>
  </div>
</section>


```python
# Mesures de dispersion du prix

afficher_tableau(df_analyse[["price"]].describe())
```

```python
# Calcul des seuils IQR

q1 = df_analyse["price"].quantile(0.25)
q3 = df_analyse["price"].quantile(0.75)
iqr = q3 - q1

seuil_bas_iqr = q1 - 1.5 * iqr
seuil_haut_iqr = q3 + 1.5 * iqr

resume_iqr = pd.DataFrame({
    "Indicateur": ["Q1", "Q3", "IQR", "Seuil bas", "Seuil haut"],
    "Valeur": [q1, q3, iqr, seuil_bas_iqr, seuil_haut_iqr]
})

afficher_tableau(resume_iqr)
```

```python
# Identification des outliers de prix selon l'IQR

outliers_iqr_prix = df_analyse[
    (df_analyse["price"] < seuil_bas_iqr) | (df_analyse["price"] > seuil_haut_iqr)
].copy().sort_values("price", ascending=False)

proportion_outliers_iqr = len(outliers_iqr_prix) / len(df_analyse) * 100 if len(df_analyse) > 0 else 0

print(f"Nombre d'articles outliers selon l'IQR : {len(outliers_iqr_prix)}")
print(f"Proportion du catalogue : {proportion_outliers_iqr:.2f} %")

afficher_tableau(outliers_iqr_prix[[col for col in ["product_id", "sku", "post_title", "price", "total_sales"] if col in outliers_iqr_prix.columns]])
```

**Conclusion sur les prix**

Les prix identifiés comme atypiques doivent être analysés avec prudence : dans un catalogue de vins, un prix élevé peut correspondre à un produit premium et ne constitue pas automatiquement une erreur. Ces produits sont donc isolés pour contrôle métier plutôt que supprimés.

```python
# Calcul du nombre de mois de stock
# Hypothèse : total_sales correspond aux ventes du mois d'octobre.
# mois_stock = stock_quantity / total_sales uniquement si les ventes sont strictement positives.

masque_mois_stock = (
    df_analyse["donnee_stock_valide"]
    & df_analyse["donnee_vente_valide"]
    & (df_analyse["total_sales"] > 0)
)

df_analyse["mois_stock_calculable"] = masque_mois_stock
df_analyse["mois_stock"] = 0.0
df_analyse.loc[masque_mois_stock, "mois_stock"] = (
    df_analyse.loc[masque_mois_stock, "stock_quantity"]
    / df_analyse.loc[masque_mois_stock, "total_sales"]
)

# Produits avec le plus de mois de stock, en excluant les produits sans ventes positives
flop20_mois_stock = df_analyse[df_analyse["mois_stock_calculable"]].sort_values("mois_stock", ascending=False).head(20)

print(f"Produits exclus du calcul des mois de stock : {int((~df_analyse['mois_stock_calculable']).sum())}")
afficher_tableau(flop20_mois_stock[[col for col in ["product_id", "sku", "post_title", "stock_quantity", "total_sales", "mois_stock"] if col in flop20_mois_stock.columns]])
```

```python
# Valorisation des stocks en euros
# Valorisation au coût d'achat et valorisation au prix de vente TTC

df_analyse["valorisation_stock_achat"] = df_analyse["stock_quantity"] * df_analyse["purchase_price"]
df_analyse["valorisation_stock_vente_ttc"] = df_analyse["stock_quantity"] * df_analyse["price"]

valorisation_stock_achat_total = df_analyse["valorisation_stock_achat"].sum()
valorisation_stock_vente_total = df_analyse["valorisation_stock_vente_ttc"].sum()
nombre_produits_stock_total = df_analyse["stock_quantity"].sum()

resume_stock = pd.DataFrame({
    "Indicateur": [
        "Nombre total d'unités en stock",
        "Valorisation du stock au prix d'achat",
        "Valorisation du stock au prix de vente TTC"
    ],
    "Valeur": [
        nombre_produits_stock_total,
        valorisation_stock_achat_total,
        valorisation_stock_vente_total
    ]
})
resume_stock["Unité"] = ["unités", "€", "€"]
resume_stock["Valeur formatée"] = resume_stock.apply(lambda ligne: (f"{ligne['Valeur']:,.2f} €".replace(",", " ").replace(".", ",") if ligne["Unité"] == "€" else f"{ligne['Valeur']:,.2f} %".replace(",", " ").replace(".", ",") if ligne["Unité"] == "%" else f"{ligne['Valeur']:,.0f}".replace(",", " ")), axis=1)

afficher_tableau(resume_stock)
```

    5.4 — Analyse du taux de marge
  </h3>
  <div style="color:#1f2937; font-family:Arial, Helvetica, sans-serif; font-size:1rem;">
<p style="margin:8px 0; line-height:1.6;">Calcul du prix HT, de la marge unitaire et du taux de marge.</p>

```python
# Calcul de la marge
# Hypothèse : les prix du fichier ERP sont des prix TTC et la TVA appliquée est de 20 %.

TAUX_TVA = 0.20

df_analyse["prix_ht"] = df_analyse["price"] / (1 + TAUX_TVA)
df_analyse["marge_unitaire"] = df_analyse["prix_ht"] - df_analyse["purchase_price"]
df_analyse["donnee_marge_valide"] = (
    df_analyse["donnee_prix_valide"]
    & df_analyse["purchase_price"].notna()
    & (df_analyse["purchase_price"] > 0)
)
df_analyse["taux_marge"] = 0.0
df_analyse.loc[df_analyse["donnee_marge_valide"], "taux_marge"] = (
    df_analyse.loc[df_analyse["donnee_marge_valide"], "marge_unitaire"]
    / df_analyse.loc[df_analyse["donnee_marge_valide"], "purchase_price"]
    * 100
)
df_analyse["marge_totale"] = np.where(
    df_analyse["donnee_marge_valide"] & df_analyse["donnee_vente_valide"],
    df_analyse["marge_unitaire"] * df_analyse["total_sales"],
    0
)

serie_taux_marge_valide = df_analyse.loc[df_analyse["donnee_marge_valide"], "taux_marge"]

resume_marge = pd.DataFrame({
    "Indicateur": [
        "Lignes avec marge calculable",
        "Lignes exclues du calcul de marge",
        "Taux de marge minimum",
        "Taux de marge maximum",
        "Taux de marge moyen",
        "Marge totale estimée"
    ],
    "Valeur": [
        int(df_analyse["donnee_marge_valide"].sum()),
        int((~df_analyse["donnee_marge_valide"]).sum()),
        serie_taux_marge_valide.min() if not serie_taux_marge_valide.empty else 0,
        serie_taux_marge_valide.max() if not serie_taux_marge_valide.empty else 0,
        serie_taux_marge_valide.mean() if not serie_taux_marge_valide.empty else 0,
        df_analyse["marge_totale"].sum()
    ]
})
resume_marge["Unité"] = ["nombre", "nombre", "%", "%", "%", "€"]
resume_marge["Valeur formatée"] = resume_marge.apply(lambda ligne: (f"{ligne['Valeur']:,.2f} €".replace(",", " ").replace(".", ",") if ligne["Unité"] == "€" else f"{ligne['Valeur']:,.2f} %".replace(",", " ").replace(".", ",") if ligne["Unité"] == "%" else f"{ligne['Valeur']:,.0f}".replace(",", " ")), axis=1)

afficher_tableau(resume_marge)
```

<p style="margin:8px 0; line-height:1.6;">Mesure des relations entre prix, prix d’achat, stock, ventes, CA, prix HT et taux de marge.</p>

```python
# Matrice de corrélation sur les variables quantitatives principales

variables_correlation = [
    "price", "purchase_price", "prix_ht", "taux_marge", "marge_unitaire",
    "stock_quantity", "total_sales", "ca_par_article", "mois_stock", "valorisation_stock_achat"
]
variables_correlation = [colonne for colonne in variables_correlation if colonne in df_analyse.columns]

df_correlation = df_analyse[variables_correlation].replace([np.inf, -np.inf], np.nan).copy()
df_correlation = df_correlation.dropna(axis=1, how="all")
df_correlation = df_correlation[[colonne for colonne in df_correlation.columns if df_correlation[colonne].nunique(dropna=True) > 1]]
df_correlation = df_correlation.dropna(how="any")

if df_correlation.shape[1] >= 2 and len(df_correlation) >= 2:
    matrice_correlation = df_correlation.corr(numeric_only=True)
    afficher_tableau(matrice_correlation, titre="Matrice de corrélation")
else:
    matrice_correlation = pd.DataFrame()
    afficher_message("Matrice de corrélation non calculable : données numériques insuffisantes après exclusion des NaN/inf.", "warning")
```

```python
# Heatmap de corrélation

if not matrice_correlation.empty:
    plt.figure(figsize=(12, 8))
    mask = np.triu(np.ones_like(matrice_correlation, dtype=bool))
    sns.heatmap(
        matrice_correlation,
        mask=mask,
        annot=True,
        fmt=".2f",
        cmap="coolwarm",
        center=0,
        linewidths=0.5
    )
    plt.title("Matrice de corrélation des variables quantitatives")
    plt.show()
else:
    afficher_message("Heatmap non affichée car la matrice de corrélation est vide.", "warning")
```

```python
# Extraction des corrélations les plus fortes en valeur absolue

if not matrice_correlation.empty:
    corr_pairs = (
        matrice_correlation
        .where(~np.eye(matrice_correlation.shape[0], dtype=bool))
        .stack()
        .reset_index()
    )
    corr_pairs.columns = ["Variable 1", "Variable 2", "Corrélation"]
    corr_pairs["Corrélation absolue"] = corr_pairs["Corrélation"].abs()

    corr_pairs_fortes = corr_pairs.sort_values("Corrélation absolue", ascending=False).drop_duplicates(
        subset=["Corrélation absolue", "Corrélation"]
    ).head(15)
else:
    corr_pairs_fortes = pd.DataFrame(columns=["Variable 1", "Variable 2", "Corrélation", "Corrélation absolue"])

afficher_tableau(corr_pairs_fortes)
```

**Conclusion sur les corrélations**

Les corrélations permettent d’identifier les relations statistiques entre variables, mais elles ne prouvent pas une causalité. Les résultats doivent donc être interprétés avec prudence et complétés par une analyse métier.

```python
# ============================================================
# Génération du dashboard HTML livrable
# Données dynamiques issues des DataFrames calculés dans le notebook
# Compatible Google Colab
# ============================================================

from pathlib import Path
from datetime import datetime
import html
import re
import base64
from io import BytesIO

DASHBOARD_DIR = Path("exports") / "tableaux"
DASHBOARD_DIR.mkdir(parents=True, exist_ok=True)
DASHBOARD_PATH = DASHBOARD_DIR / "dashboard_livrable.html"


def _format_nombre(valeur, decimales=0, suffixe=""):
    """Formate un indicateur numérique en notation française."""
    try:
        nombre = float(valeur)
    except (TypeError, ValueError):
        return "n/a"
    if pd.isna(nombre):
        return "n/a"
    texte = f"{nombre:,.{decimales}f}".replace(",", " ").replace(".", ",")
    return f"{texte}{suffixe}"


def _trouver_colonne(df, candidats):
    """Retourne la première colonne existante parmi une liste de noms possibles."""
    if df is None:
        return None
    correspondances = {str(col).lower().strip(): col for col in df.columns}
    for candidat in candidats:
        cle = candidat.lower().strip()
        if cle in correspondances:
            return correspondances[cle]
    for col in df.columns:
        col_norm = str(col).lower().strip()
        if any(candidat.lower().strip() in col_norm for candidat in candidats):
            return col
    return None



def _graphique_barres_html(df, titre, colonne_valeur, colonne_libelle=None, max_lignes=10, couleur="#33A5B6"):
    """Génère un graphique en barres horizontal encodé en base64 pour le dashboard."""
    plt_obj = globals().get("plt")
    if plt_obj is None or df is None or colonne_valeur is None or colonne_valeur not in df.columns or len(df) == 0:
        return f"<section class='panel chart'><h2>{html.escape(titre)}</h2><p class='empty'>Graphique non disponible.</p></section>"

    data = df[[colonne_valeur] + ([colonne_libelle] if colonne_libelle in df.columns else [])].copy()
    data[colonne_valeur] = pd.to_numeric(data[colonne_valeur], errors="coerce")
    data = data.dropna(subset=[colonne_valeur]).sort_values(colonne_valeur, ascending=True).tail(max_lignes)
    if data.empty:
        return f"<section class='panel chart'><h2>{html.escape(titre)}</h2><p class='empty'>Graphique non disponible.</p></section>"

    labels = data[colonne_libelle].astype(str) if colonne_libelle in data.columns else data.index.astype(str)
    labels = labels.str.slice(0, 52)
    fig, ax = plt_obj.subplots(figsize=(10, max(4, len(data) * 0.45)))
    ax.barh(labels, data[colonne_valeur], color=couleur)
    ax.set_title(titre)
    ax.grid(axis="x", alpha=0.25)
    ax.set_xlabel(str(colonne_valeur))
    fig.tight_layout()

    buffer = BytesIO()
    fig.savefig(buffer, format="png", dpi=140, bbox_inches="tight")
    plt_obj.close(fig)
    image_base64 = base64.b64encode(buffer.getvalue()).decode("ascii")
    return (
        f"<section class='panel chart'><h2>{html.escape(titre)}</h2>"
        f"<img src='data:image/png;base64,{image_base64}' alt='{html.escape(titre)}'></section>"
    )

def _table_html(df, titre, max_lignes=10):
    """Transforme un DataFrame en bloc HTML sécurisé."""
    if df is None or len(df) == 0:
        return f"<section class='panel'><h2>{html.escape(titre)}</h2><p class='empty'>Aucune donnée disponible.</p></section>"
    extrait = df.head(max_lignes).copy()
    return (
        f"<section class='panel'><h2>{html.escape(titre)}</h2>"
        + extrait.to_html(index=False, border=0, classes='data-table', escape=True)
        + "</section>"
    )

base_dashboard = globals().get("df_analyse", globals().get("df_merge", None))
journal_dashboard = globals().get("journal_anomalies", None)
recommandations_dashboard = globals().get("recommandations_erp", None)

if base_dashboard is None or not hasattr(base_dashboard, "columns"):
    raise ValueError("Le dashboard nécessite df_analyse ou df_merge. Relance les cellules de chargement, nettoyage et consolidation avant cette cellule.")

col_ca = _trouver_colonne(base_dashboard, ["ca_par_article", "chiffre_affaires", "chiffre d'affaires", "ca"])
if col_ca is None and {"price", "total_sales"}.issubset(base_dashboard.columns):
    col_ca = "ca_par_article"
    base_dashboard = base_dashboard.copy()
    base_dashboard[col_ca] = base_dashboard["price"] * base_dashboard["total_sales"]
col_prix = _trouver_colonne(base_dashboard, ["price", "prix"])
col_ventes = _trouver_colonne(base_dashboard, ["total_sales", "sales", "ventes", "quantite_vendue", "quantité vendue"])
col_stock = _trouver_colonne(base_dashboard, ["stock_quantity", "stock", "quantite_stock", "quantité stock"])
col_marge = _trouver_colonne(base_dashboard, ["marge_totale", "marge totale", "marge", "margin"])
col_sku = _trouver_colonne(base_dashboard, ["sku", "id_web", "product_id"])
col_nom = _trouver_colonne(base_dashboard, ["post_title", "nom", "product_name", "libelle", "libellé"])

ca_total = base_dashboard[col_ca].sum() if col_ca else None
ventes_total = base_dashboard[col_ventes].sum() if col_ventes else None
stock_total = base_dashboard[col_stock].sum() if col_stock else None
prix_median = base_dashboard[col_prix].median() if col_prix else None
marge_total = base_dashboard[col_marge].sum() if col_marge else None
nb_produits = len(base_dashboard)
nb_anomalies = len(journal_dashboard) if journal_dashboard is not None else 0

tri_ca = base_dashboard.sort_values(col_ca, ascending=False) if col_ca else base_dashboard
colonnes_top = []
for col in [col_sku, col_nom, col_ca, col_ventes, col_stock, col_prix, col_marge]:
    if col is not None and col not in colonnes_top:
        colonnes_top.append(col)
top_ca = tri_ca[colonnes_top].head(10) if colonnes_top else tri_ca.head(10)
stock_prioritaire = base_dashboard.sort_values(col_stock, ascending=False)[colonnes_top].head(10) if col_stock and colonnes_top else None
prix_prioritaire = base_dashboard.sort_values(col_prix, ascending=False)[colonnes_top].head(10) if col_prix and colonnes_top else None

kpis = [
    ("Chiffre d'affaires", _format_nombre(ca_total, 2, " €") if ca_total is not None else "n/a", "Somme recalculée depuis la base d'analyse."),
    ("Produits analysés", _format_nombre(nb_produits, 0), "Nombre de lignes de la base consolidée utilisée."),
    ("Ventes", _format_nombre(ventes_total, 0) if ventes_total is not None else "n/a", "Quantités vendues ou ventes web selon les colonnes disponibles."),
    ("Stock total", _format_nombre(stock_total, 0) if stock_total is not None else "n/a", "Stock cumulé issu des données réelles du notebook."),
    ("Prix médian", _format_nombre(prix_median, 2, " €") if prix_median is not None else "n/a", "Médiane des prix détectés dans la base."),
    ("Anomalies", _format_nombre(nb_anomalies, 0), "Lignes du journal d'anomalies généré par l'analyse."),
]
if marge_total is not None:
    kpis.append(("Marge totale", _format_nombre(marge_total, 2, " €"), "Marge liée aux ventes : somme de marge_unitaire × total_sales."))

kpi_html = "".join(
    f"<article class='kpi'><span>{html.escape(titre)}</span><strong>{html.escape(valeur)}</strong><small>{html.escape(note)}</small></article>"
    for titre, valeur, note in kpis
)

sections_graphiques = "".join([
    _graphique_barres_html(tri_ca, "Top produits par chiffre d'affaires", col_ca, col_nom or col_sku, 10, "#33A5B6"),
    _graphique_barres_html(base_dashboard.sort_values(col_ventes, ascending=False) if col_ventes else None, "Top produits par ventes", col_ventes, col_nom or col_sku, 10, "#1E7D8B"),
    _graphique_barres_html(base_dashboard.sort_values(col_stock, ascending=False) if col_stock else None, "Stocks les plus élevés", col_stock, col_nom or col_sku, 10, "#F2B134"),
    _graphique_barres_html(base_dashboard.sort_values(col_marge, ascending=False) if col_marge else None, "Top marges", col_marge, col_nom or col_sku, 10, "#607D8B"),
])

sections_tableaux = "".join([
    _table_html(top_ca, "Top produits par chiffre d'affaires", 10),
    _table_html(stock_prioritaire, "Stocks prioritaires", 10),
    _table_html(prix_prioritaire, "Prix les plus élevés", 10),
    _table_html(journal_dashboard, "Journal des anomalies", 10),
    _table_html(recommandations_dashboard, "Recommandations ERP", 10),
])

generated_at = datetime.now().strftime("%d/%m/%Y %H:%M")
colonnes_detectees = []
for col in [col_ca, col_ventes, col_stock, col_prix, col_marge]:
    if col is not None and col not in colonnes_detectees:
        colonnes_detectees.append(col)
colonnes_utilisees = ", ".join(html.escape(str(c)) for c in colonnes_detectees) or "colonnes disponibles automatiquement détectées"

dashboard_html = f"""<!doctype html>
<html lang="fr">
<head>
  <meta charset="utf-8">
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>Dashboard livrable BottleNeck</title>
  <style>
    :root {{ --primary:#33A5B6; --dark:#111827; --muted:#607D8B; --line:#dbe7ee; --soft:#f4fafb; }}
    * {{ box-sizing:border-box; }}
    body {{ margin:0; font-family:Inter,Segoe UI,Arial,sans-serif; color:#1f2937; background:#eef4f8; }}
    .hero {{ padding:34px clamp(18px,4vw,54px); color:#fff; background:linear-gradient(135deg,#33A5B6 0%,#1E7D8B 58%,#111827 100%); }}
    .hero h1 {{ margin:0 0 8px; font-size:clamp(1.7rem,4vw,3rem); }}
    .hero p {{ margin:0; max-width:1050px; line-height:1.6; color:#eaf9fc; }}
    .nav {{ position:sticky; top:0; z-index:5; display:flex; gap:10px; flex-wrap:wrap; padding:12px clamp(18px,4vw,54px); background:#ffffffee; border-bottom:1px solid var(--line); backdrop-filter:blur(8px); }}
    .nav a {{ color:#1E7D8B; text-decoration:none; font-weight:800; padding:8px 12px; border-radius:999px; background:#eaf9fc; }}
    main {{ padding:22px clamp(18px,4vw,54px) 44px; }}
    .kpi-grid {{ display:grid; grid-template-columns:repeat(auto-fit,minmax(210px,1fr)); gap:16px; margin:0 0 24px; }}
    .kpi, .panel {{ background:#fff; border:1px solid var(--line); border-radius:18px; box-shadow:0 12px 30px rgba(17,24,39,.08); }}
    .kpi {{ padding:18px; }}
    .kpi span {{ color:var(--muted); font-weight:800; text-transform:uppercase; letter-spacing:.06em; font-size:.78rem; }}
    .kpi strong {{ display:block; margin:8px 0; color:#111827; font-size:clamp(1.45rem,3vw,2.25rem); }}
    .kpi small {{ color:#607D8B; line-height:1.45; }}
    .panel {{ margin:18px 0; padding:18px; overflow:auto; }}
    .chart-grid {{ display:grid; grid-template-columns:1fr; gap:30px; margin:0 0 24px; }}
    .chart img {{ width:100%; max-width:1480px; min-height:520px; max-height:none; height:auto; object-fit:contain; display:block; margin:0 auto; border-radius:12px; border:1px solid var(--line); }}
    .panel h2 {{ margin:0 0 14px; color:#111827; }}
    .data-table {{ width:100%; border-collapse:collapse; font-size:.92rem; }}
    .data-table th {{ text-align:left; background:#eaf9fc; color:#1E7D8B; }}
    .data-table th, .data-table td {{ padding:10px 11px; border-bottom:1px solid var(--line); vertical-align:top; }}
    .empty {{ color:#607D8B; }}
    .note {{ background:#fff; border-left:6px solid #33A5B6; border-radius:14px; padding:14px 16px; margin:0 0 20px; line-height:1.55; }}
    footer {{ color:#607D8B; text-align:center; padding:24px; }}
  </style>
</head>
<body>
  <header class="hero" id="top"><h1>Dashboard livrable BottleNeck</h1><p>Vue exécutive générée automatiquement depuis les DataFrames du notebook. Les KPI, classements, anomalies et recommandations sont recalculés à chaque exécution afin de rester alignés avec les analyses réelles.</p></header>
  <nav class="nav"><a href="#kpi">KPI</a><a href="#graphes">Graphiques</a><a href="#tableaux">Tableaux</a><a href="#actions">Actions</a><a href="#top">↑ Haut</a></nav>
  <main>
    <p class="note"><strong>Source dynamique :</strong> dashboard généré le {generated_at} depuis <code>df_analyse</code> ou <code>df_merge</code>. Colonnes détectées : {colonnes_utilisees}.</p>
    <section id="kpi" class="kpi-grid">{kpi_html}</section>
    <section id="graphes" class="chart-grid">{sections_graphiques}</section>
    <section id="tableaux">{sections_tableaux}</section>
    <section id="actions" class="panel"><h2>Synthèse pour décision</h2><p>Prioriser le contrôle des prix atypiques, la correction des clés de jointure, le traitement du stock dormant et l'industrialisation des contrôles ERP/Web afin de fiabiliser les prochains exports.</p></section>
  </main>
  <footer>BottleNeck — livrable notebook + dashboard HTML dynamique</footer>
</body>
</html>
"""

DASHBOARD_PATH.write_text(dashboard_html, encoding="utf-8")
if "_enregistrer_export_html" in globals():
    _enregistrer_export_html(
        DASHBOARD_PATH,
        "Dashboard livrable BottleNeck",
        "Accueil / Dashboard",
        "Exports clés",
        "Vue exécutive avec KPI, graphiques et tableaux prioritaires.",
        None,
    )
print("Dashboard HTML généré :")
print(DASHBOARD_PATH.resolve())

```
