```python
# Vérifier s'l y a des lignes entièrement dupliquées

nb_lignes_dupliquees = df_erp.duplicated().sum()

print(f"Nombre de lignes entièrement dupliquées : {nb_lignes_dupliquees}")
```

```python
# Afficher les valeurs distinctes de la colonne stock_status
# À quelle(s) autre(s) colonne(s) sont-elles liées ?

df_erp["stock_status"].value_counts(dropna=False)
```

```python
# Croisement entre stock_status et stock_quantity
# Objectif : vérifier si le statut de stock est cohérent avec la quantité en stock

pd.crosstab(
    df_erp["stock_status"],
    df_erp["stock_quantity"] == 0,
    margins=True
)
```

```python
#Création d'une colonne "stock_status_2"
#La valeur de cette deuxième colonne sera fonction de la valeur dans la colonne "stock_quantity"
#Si la valeur de la colonne "stock_quantity" est nulle, renseigner "outofstock" sinon mettre "instock"

df_erp["stock_status_2"] = np.where(
    df_erp["stock_quantity"] == 0,
    "outofstock",
    "instock"
)

df_erp[["product_id", "stock_quantity", "stock_status", "stock_status_2"]].head()
```

```python
#Vérifions que les 2 colonnes sont identiques:
#Les 2 colonnes sont strictement identiques si les valeurs de chaque ligne sont strictement identiques 2 à 2
#La comparaison de 2 colonnes peut se réaliser simplement avec l'instruction ci-dessous:
#df_erp["stock_status"] == df_erp["stock_status_2"]

#Le résultat est l'affichage de True ou False pour chacune des lignes du dataset
#C'est un bon début, mais difficile à exploiter

# Vérification ligne par ligne de la cohérence entre stock_status et stock_status_2

comparaison_stock_status = df_erp["stock_status"] == df_erp["stock_status_2"]

comparaison_stock_status.value_counts()
```

```python
#Mais il est possible de synthétiser ce résultat en effectuant la somme de cette colonne:
#True vaut 1 et False 0
#Nous devrions obtenir la somme de 824 qui correspond au nombre de lignes dans ce dataset

nb_statuts_identiques = comparaison_stock_status.sum()
nb_ecarts_statut = (~comparaison_stock_status).sum()

print(f"Nombre de lignes avec un statut cohérent : {nb_statuts_identiques}")
print(f"Nombre de lignes avec un écart de statut : {nb_ecarts_statut}")
```

```python
#Si les colonnes ne sont absolument pas identiques ligne à ligne alors identifier la ligne en écart
##Dans ce cas je vous donne ce lien pour apprendre à réaliser des filtres dans Pandas:
##https://bitbucket.org/hrojas/learn-pandas/src/master/
##Lesson 3

df_erp.loc[
    df_erp["stock_status"] != df_erp["stock_status_2"],
    ["product_id", "stock_quantity", "stock_status", "stock_status_2"]
]
```

```python
#Corriger la ou les données incohérentes
# Création d'un dataframe contenant les incohérences de stock
anomalies_stock_status = df_erp.loc[
    df_erp["stock_status"] != df_erp["stock_status_2"],
    ["product_id", "stock_quantity", "stock_status", "stock_status_2"]
]

#Vérification en utilisant le même code que plus haut pour afficher les problèmes
anomalies_stock_status
```

### Observation

Le contrôle entre `stock_quantity` et `stock_status` met en évidence plusieurs incohérences.

La règle utilisée est la suivante :
- si `stock_quantity` est égal à 0, le produit devrait être `outofstock` ;
- si `stock_quantity` est différent de 0, le produit devrait être `instock`.

Les lignes en écart doivent être isolées afin de ne pas fausser l’analyse du stock.
La correction peut être proposée, mais elle ne doit pas être appliquée sans règle métier validée.
